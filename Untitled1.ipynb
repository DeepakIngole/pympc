```python
import numpy as np
import matplotlib.pyplot as plt
from pympc.geometry.polytope import Polytope
from pympc.geometry.inner_approximation_polytope_projection import InnerApproximationOfPolytopeProjection, point_in_convex_hull
```

```python
points = [np.random.randn(2, 1) for i in range(5)]
x = np.random.randn(2, 1)/5.
plt.scatter(x[0,0], x[1,0], color='r')
for v in points:
    plt.scatter(v[0,0], v[1,0], color='g')
plt.show()
print point_in_convex_hull(x, points)
```

```python
n_var = 5
n_cons = 200
A = np.random.randn(n_cons, n_var)
b = np.random.rand(n_cons, 1)
p0 = Polytope(A, b)
p0.assemble()
p0.plot(alpha=.2)
plt.show()
```

```python
app = InnerApproximationOfPolytopeProjection(A, b, [0,1])

for i in range(10):
    x = np.random.randn(2, 1)/50.
    p0.plot(alpha=.2)
    plt.scatter(x[0,0], x[1,0], color='r')
    plt.show()
    raw_input("Press Enter to continue...")
    app.include_point(x)
    plt.scatter(x[0,0], x[1,0], color='r')
    for v in app.vertices:
        plt.scatter(v[0,0], v[1,0], color='g')
    p0.plot(alpha=.2)
    plt.show()
```
